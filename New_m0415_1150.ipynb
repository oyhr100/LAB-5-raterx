# **basic settings**

```python
import os
import pandas as pd
import numpy as np

from itables import init_notebook_mode
init_notebook_mode(all_interactive=True)
```

```python
df_crime = (
    pd.read_csv('df_crime.csv').drop(columns = 'Block')
)

df_crime = df_crime[df_crime['psa']!='A']
df_crime.replace('nan', np.nan, inplace=True)
df_crime
```

```python
df_crime_days = df_crime.drop(columns = 'Category')
df_crime_days
```

```python
import pandas as pd
from itertools import product


data_ranges = {}

for i in df_crime_days.columns:
    data_ranges[i] = df_crime_days[i].unique()

print(data_ranges)
```

```python
import itertools

# 
all_combinations = list(itertools.product(*data_ranges.values()))

# 
complement = [combo for combo in all_combinations if combo not in df_crime_days]

complement
```

```python
import random

n=df_crime_days.shape[0]
complement = [random.choice(complement) for _ in range(3*n)]
```

```python
column_names = list(data_ranges.keys())

#  NumPy 
complement_array = np.array(complement)

#  DataFrame
complement_df = pd.DataFrame(complement_array, columns=column_names)
complement_df
```

```python
merged_df = pd.concat([df_crime, complement_df], ignore_index=True)

merged_df['Category'] = merged_df['Category'].fillna("No Crime")
```

```python
merged_df['Crime'] = (merged_df['Category'] != "No Crime").astype(int)
merged_df = merged_df.drop(columns = 'Category')

#  'nan'  NaN 0
merged_df.replace('nan', np.nan, inplace=True)  #  'nan' 
merged_df = merged_df.dropna().fillna(0).astype(int)
merged_df
```

```python
df_climate = pd.read_excel('climate_philly.xlsx')

print(df_climate.head())
```

```python
#  'DATE' 
df_climate['Year'] = df_climate['DATE'].dt.year
df_climate['Month'] = df_climate['DATE'].dt.month
df_climate['Day'] = df_climate['DATE'].dt.day

#  'DATE' 
df_climate.drop(columns=['DATE'], inplace=True)

#  DataFrame
merged_df = pd.merge(merged_df, df_climate, on=['Day', 'Month', 'Year'], how='inner')
merged_df
```

```python
# 
missing_values = merged_df.isnull().sum()
print("\n", missing_values)

# 
print("\n")
print(missing_values[missing_values > 0])

# 
merged_df.dropna(inplace=True)
```

```python
df_last = pd.get_dummies(merged_df, columns=['dc_dist','psa','Day','DayOfWeek',"Month",'Year','Hour'], drop_first=True)
df_last
```

```python
#  'PRCP''TMAX'  'TMIN' 
cols_to_convert = [col for col in df_last.columns if col not in ['PRCP', 'TMAX', 'TMIN']]

```

```python
df_last[cols_to_convert] = df_last[cols_to_convert].astype(int)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X, y = df_last.drop(columns = 'Crime'), df_last['Crime']

# 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# 
model = LogisticRegression(max_iter=10000,solver='lbfgs')

# 
model.fit(X_train, y_train)

# 
accuracy = model.score(X_test, y_test)
print("", accuracy)
```
